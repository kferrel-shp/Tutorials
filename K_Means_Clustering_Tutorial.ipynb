A k-means clustering analysis is a machine learning technique that groups similar data point into clusters based on how many of these clusters the user, us, would like to identify. These clusters are based on their proximity in space.

In this technique, the algorithim assigns each of the data points to one of the clusters by iteratively, or repeating the process of, finding the point, known as the centroid.

The centroid is a central point that minimizes the distance between the points and the centroid. These groups are then labeled as a cluster.

[here](https://www.ibm.com/topics/k-means-clustering#:~:text=The%20k%2Dmeans%20clustering%20algorithm%20operates%20by%20categorizing%20data%20points,larger%20clusters%20with%20less%20detail.) is a article by IBM that explains in more detail what a k-means analysis is and how it works.

```python
                                             # Library imports
import sklearn
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

Step 1, define the example data. We're working with 2-D data (x,y)

```python
#500 points
num_points = 500

# generate 500 random points with x and y values between -100 and 350. Play around with these values to see how things change
data = np.random.uniform(low=-100, high=350, size=(num_points,2))
plt.scatter(data[:, 0], data[:, 1], color='blue', alpha=0.5, edgecolor='b')

#data[:,0], [:,1] just accesses the x and y values
# color = blue, this one is easy
#alpha = 0.5 makes sets the transparency of points. This helps see overlapping points
# edgecolor = 'b' makes the outline of each point black so visualization is easier
```

Step 2, Cluster identification. Set out the parameters you wish to find in the clustering

```python
# define the number of clusters(n_clusters) I want to identify (2)
kmeans = KMeans(n_clusters=2, random_state=0)

# if you want to identify a different amount of clusters just change the number following the parameter

                  #example with 3 clusters

#kmeans = KMeans(n_clusters =3, random_state=0)



                  #example with 5 clusters

#kmeans = KMeans(n_clusters =5, random_state=0)
```

step 3, fit data into the k-means model.
Essentially just loading the data

```python
#fitting the data to the model
kmeans.fit(data)
```

Step 4, set the cluster labels.

```python
#setting cluster labels
labels = kmeans.labels_

#identify the coordinates of the two centroids
centroids = kmeans.cluster_centers_

#print("Cluster labels: ", labels)
print("Centroid coordinates: ", centroids)
```

Step 5, set our plot to show the pretty results

```python
# figure size
plt.figure(figsize=(8,8))

# put each point on a scatter plot using a loop

for i in range(kmeans.n_clusters):
    # choose the points that belong to the current cluster
    cluster_points = data[labels==i]
    #plotting cluster 1 (0) and cluster 2 (1)
    # take note, if you change the amount of clusters to a value different than 2 you have to edit this section
    # "label =.." denotes which point is associated to which cluster
    plt.scatter(cluster_points[:,0], cluster_points[:,1], label = f'Cluster {i+1}')


# plot the centroids

#color is red
#marker is the shape of an x
#s=100 adjust the size of the marker
# each marker is labeled 'centroids'
plt.scatter(centroids[:,0], centroids[:,1], color= 'red', marker = 'o', s=100, label='Centroids')


# show the results
plt.title('K-Means Clustering Test Trial') # plot title
plt.xlabel('X-axis') #x-axis title
plt.ylabel('Y-axis') #y-axis title
plt.legend(bbox_to_anchor=(1,1)) #moves legend to the upper right, outside of the plot
plt.grid(True)#shows the grid lines on the plot

plt.show()#actually shows the plot
```

See how well that looks.. From this method you can identify statistically similar trends among 2 dimentional (x,y) data and can attempt to identify why certain data follow certain trends or is located within certain clusters.
